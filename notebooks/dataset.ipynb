```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder

df = pd.read_csv('../data/Dataset-applications.csv') 

# Xóa các cột không giúp ích cho việc nhận diện hành vi
cols_to_drop = ['Flow.ID', 'Source.IP', 'Source.Port', 'Destination.IP', 'Destination.Port', 'Timestamp']

df_clean = df.drop(columns=[c for c in cols_to_drop if c in df.columns])

print(f"Rest of columns: {df_clean.shape[1]}")

# Label Encoding
le = LabelEncoder()
df_clean['ProtocolName'] = le.fit_transform(df_clean['ProtocolName'])

# In ra: 0 là gì, 1 là gì...
print("Danh sách nhãn:", dict(zip(le.classes_, range(len(le.classes_)))))

# Lưu từ điển nhãn
label_dict = dict(zip(range(len(le.classes_)), le.classes_))
print("Từ điển tra cứu ngược:", label_dict)

# Hiển thị 5 dòng đầu tiên của file dữ liệu
df.head()
```
